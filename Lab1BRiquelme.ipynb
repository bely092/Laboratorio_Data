```python
# 8 Crear un nuevo dataframe SUBDF formado solamente con las
# caracteristicas del dataframe DF que tienen menos de 50 valores nulos
subdf = df.drop(['eqivalent_fips_code', 'neighbours', 'postal_code_format', 'postal_code_regex'],axis=1)
```

```python
# 9 Visualizar Cuantos valores nulos quedan en las caracteristicas del
# dataframe "subdf"
subdf.isnull().sum()
```

```python
# 10 Eliminar todas las observaciones con valores nulos del subdf
# y colocar los resultados en un nuevo dataframe de nombre: subdf1
subdf1 = subdf.dropna()
```

```python
# 11 Visualizar cuantas observaciones y caracteriticas tiene subdf1
subdf1.info()
```

```python
# 12 Verificar que no queden valores nulos en subdf1
subdf1.isnull().sum()
```

```python
# 14 Visualizar a traves de un grafico scatter la relacionn entre las columnas
# Area (eje x) y Population (eje y) de subdf1
#subdf1 = subdf1.astype(int)

fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(x = subdf1['area'],y = subdf1['population'],color="red",alpha= 0.6)

plt.xlabel("Area")
plt.ylabel("Population")

plt.show()
```

```python
# 15 Agregar una nueva columna al dataframe "subdf1" cuyo nombre sea RELACION
# y sea el resultado de dividir Population/Area
x = subdf1['area']/subdf1['population']
subdf1.insert(loc=0, column="relacion", value = x)
```

```python
# 16 Visualizar las 10 primeras observaciones de subdf1 y verificar que existe
# la nueva columna que agregamos


subdf1["continent"].unique()
subdf1.head(10)
```

```python
# Ejecute este Codigo para crear una dataframe con el codigo y nombre
# de cada unos de los continentes

Continent=["EU","AS","AF","SA","OC"]
Nombre=["EUROPA","ASIA","AFRICA","SUDAMERICA","OCEANIA"]
dataC=pd.DataFrame(
    {"Continent":Continent,
     "Nombre": Nombre})
dataC.head()
```

```python
# 17 Hacer un JOIN tipo INNER entre los dataframe subdf1 y dataC,
#  determine de que forma se podrian relacionar los mismos.
#------------------------------------------------
# test sin inner pero con mas observaciones:
#dataJ = pd.merge(subdf1, dataC, left_index=True,right_index=True ,how="outer")

#dataJ = pd.merge(subdf1, dataC, left_index=True, right_index=True ,how="inner")

dataJ = pd.merge(left=subdf1, right=dataC, how="inner", left_on="continent", right_on="Continent")


```
